```python
# 피처 데이터
feature_data = credit.drop('credit.rating', axis=1)
feature_data.head(2)

# 레이블 데이터
labels = credit['credit.rating']
labels.head(5)
```

```python
# 데이터 준비 / 분리
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature_data,
                                                    labels,
                                                    random_state=11,
                                                    stratify=labels)
```

```python
pd.Series(labels).value_counts()
pd.Series(labels).value_counts(normalize=True)

pd.Series(y_train).value_counts()
pd.Series(y_train).value_counts(normalize=True)
```

```python
# 의사결정 트리 Classifier (분류기) 생성 및 학습
from sklearn.tree import DecisionTreeClassifier

# 모델 생성
dt_clf = DecisionTreeClassifier(random_state=156)
dt_clf.fit(X_train, y_train)
```

```python
# plot_tree() 사용하여 트리 그래프 출력
from sklearn.tree import plot_tree

plt.figure(figsize=(10, 7))
plot_tree(dt_clf, filled=True, feature_names=feature_data.columns.tolist())
plt.show()
# max_depth가 설정되지 않아 어질어질하다
```

```python
# 피처 이름 확인
feature_data.columns.tolist()
```

```python
plt.figure(figsize=(10, 7))
plot_tree(dt_clf, max_depth=1, filled=True, feature_names=feature_data.columns.tolist())
plt.show()
```

```python
# 훈련에 사용된 총 샘플 수
y_train.size
```

```python
# 결정 트리 하이퍼 파라미터 튜닝1

# Classifier(분류기/모델) 생성 시 max_depth 설정
# max_depth=3 : max_depth가 3인 트리로 생성

dt_clf = DecisionTreeClassifier(max_depth=3, random_state=156)
dt_clf.fit(X_train, y_train)

plt.figure(figsize=(10, 7))
plot_tree(dt_clf, filled=True, feature_names=feature_data.columns.tolist())
plt.show()
```

```python
# 결정 트리 하이퍼 파라미터 튜닝2
# 분할 가능한 최소 샘플 수를 4로 설정 (분할 조건)
# min_samples_split=4로 설정

dt_clf = DecisionTreeClassifier(min_samples_split=4, random_state=156)
dt_clf.fit(X_train, y_train)

plt.figure(figsize=(10, 7))
plot_tree(dt_clf, filled=True, feature_names=feature_data.columns.tolist())
plt.show()
```

```python
# 결정 트리 하이퍼 파라미터 튜닝3
# 리프 노드가 될 수 있는 최소 샘플 수 설정 (분할 조건 : 분할 자격 샘플 수)
# min_samples_leaf=4로 설정

dt_clf = DecisionTreeClassifier(min_samples_split=4, random_state=156)
dt_clf.fit(X_train, y_train)

plt.figure(figsize=(10, 7))
plot_tree(dt_clf, filled=True, feature_names=feature_data.columns.tolist())
plt.show()
```

```python
# 분류기(dt_clf)의 feature_importance_ 속성
```

```python
# 피처별 중요도
dt_clf.feature_importances_

# feature 별 importance 맵핑 : zip() 사용
for name, value in zip(feature_data.columns.tolist(), dt_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))
```

```python
import seaborn as sns
# feature_importances_를 컬럼 별로 시각화
sns.barplot(x=dt_clf.feature_importances_, y=feature_data.columns.tolist())
```
